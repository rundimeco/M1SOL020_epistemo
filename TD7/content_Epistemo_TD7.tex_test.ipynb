```python
from nltk import agreement
annotateur1 = ["DET", "NOM", "ADJ", "ADJ", "ADV", "DET","ADJ", "ADV"]
annotateur2 = ["DET", "NOM", "ADJ", "NOM", "ADV", "NOM","ADV", "ADV"]
annotateurs = [annotateur1, annotateur2]
donnees = []
for i in range(len(annotateurs)):
  for j in range(len(annotateur1)):
    donnees.append([str(i), str(j), annotateurs[i][j]])
ratingtask = agreement.AnnotationTask(data=donnees)
print("kappa " +str(ratingtask.kappa()))
print("fleiss " + str(ratingtask.multi_kappa()))
print("alpha " +str(ratingtask.alpha()))
print("scotts " + str(ratingtask.pi()))
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
annotateur1 = ["DET", "NOM", "ADJ", "ADJ", "ADV", "DET","ADJ", "ADV"]
annotateur2 = ["DET", "NOM", "ADJ", "NOM", "ADV", "NOM","ADV", "ADV"]

# les classes c'est l'ensemble des étiquettes possibles
# ou on peut les restreindre (comme ici) aux étiquettes vues chez les deux annotateurs
# set() les donne sans doublons

classes = sorted(list(set(annotateur1+annotateur2)))
#Edit du 5/12/2019 : il faut faire un sorted() de ça pour pouvoir les mettre dans l'ordre
# En effet, par défaut dans la matrice de confusion les classes sont présentées en ordre lexicographique
# la valeur donnée aux labels xtick et ytick n'a cun influence sur les données présentées
#c'est juste de l'affichage
# voir exemple puis contre-exemple ci-dessous
```

```python

cm = confusion_matrix(annotateur1, annotateur2)

fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(cm, annot=True, xticklabels=classes, yticklabels=classes, cmap=plt.cm.Reds )
plt.ylim([len(classes), 0])
print("Matrice correcte")
plt.show()
```

```python
## Ancien code avec une erreur sur les étiquettes
classes = list(set(annotateur1+annotateur2))

cm = confusion_matrix(annotateur1, annotateur2)

fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(cm, annot=True, xticklabels=classes, yticklabels=classes, cmap=plt.cm.Reds )
plt.ylim([len(classes), 0])
print("Matrice incorrecte : les données sont aux mêmes endroits mais ne correspondent pas aux labels")
plt.show()
```
